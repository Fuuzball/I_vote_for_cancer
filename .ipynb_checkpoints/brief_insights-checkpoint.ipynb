```python
import os, math
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
for f in os.listdir('./data'):
    size_bytes = round(os.path.getsize('./data/' + f)/1000, 2)
    size_name = ["KB", "MB"]
    i = int(math.floor(math.log(size_bytes, 1024)))
    p = math.pow(1024, i)
    s = round(size_bytes / p, 2)
    print(f.ljust(20) + str(s).ljust(7) + size_name[i])
```

```python
train_variants_df = pd.read_csv("./data/training_variants")
test_variants_df = pd.read_csv("./data/test_variants")
train_text_df = pd.read_csv("./data/training_text", sep="\|\|", engine="python", skiprows=1, names=["ID", "Text"])
test_text_df = pd.read_csv("./data/test_text", sep="\|\|", engine="python", skiprows=1, names=["ID", "Text"])
print("Train Variant".ljust(15), train_variants_df.shape)
print("Train Text".ljust(15), train_text_df.shape)
print("Test Variant".ljust(15), test_variants_df.shape)
print("Test Text".ljust(15), test_text_df.shape)
```

```python
plt.figure(figsize = (12,8))
sns.countplot(x='Class', data=train_variants_df, palette='Reds_d')
plt.ylabel('Frequency')
plt.xlabel('Class')
plt.title('Distribution of mutation classes')
plt.show()
```

```python
gene_group = train_variants_df.groupby('Gene').size()
print(gene_group.sort_values(ascending=False)[:10])
print(gene_group.sort_values(ascending=True)[:10])
```

```python
fig, ax = plt.subplots(ncols=3, nrows=3, figsize=(15,15))

for i in range(3):
    for j in range(3):
        gene_count_grp = train_variants_df[train_variants_df['Class'] == (i*3 + j) + 1].groupby('Gene')['ID'].count().reset_index()
        gene_top_7 = gene_count_grp.sort_values('ID', ascending=False)[:7]
        sns.barplot(x="Gene", y="ID", data=gene_top_7, ax=ax[i][j])
```

```python
train_text_df['Text_count'] = train_text_df['Text'].apply(lambda x:len(x.split()))
```

```python
train_full = train_variants_df.merge(train_text_df, how = 'inner', left_on='ID', right_on='ID')
test_full = test_variants_df.merge(test_text_df, how = 'inner', left_on='ID', right_on='ID')
```

```python
train_full['Var in Text'] = train_full.apply(lambda x : x['Text'].count(x['Variation']), axis=1)
test_full['Var in Text'] = test_full.apply(lambda x : x['Text'].count(x['Variation']), axis=1)
```

```python
print(train_full['Var in Text'].mean())
print(test_full['Var in Text'].mean())
```

```python
print(train_full['Var in Text'][train_full['Var in Text'] > 0].count()/train_full['Var in Text'].count())
print(test_full['Var in Text'][test_full['Var in Text'] > 0].count()/test_full['Var in Text'].count())
```

```python
total_train_text = ""
for text in train_full['Text']:
    total_train_text += text

total_test_text = ""
for text in test_full['Text']:
    total_test_text += text

```

```python
train_full.sample(5)
```

```python
train_full['in Any Text'] = train_full.apply(lambda x : x['Variation'] in total_train_text, axis=1)
test_full['in Any Text'] = test_full.apply(lambda x : x['Variation'] in total_train_text, axis=1)
```

```python
class_text = ["",] * 9

for class_i in range(9):
    for text in train_full['Text'][train_full['Class'] == class_i + 1]:
        class_text[class_i] += text
```

```python
def var_in_class_text(row):
    var = row['Variation']
    class_ = row['Class']
    for text in train_full['Text'][train_full['Class'] == class_]:
        if var in text:
            return True
            break
    else:
        return False
        
```

```python
train_full.head()
```

```python
train_full['in Class Text'] = train_full.apply(var_in_class_text, axis=1)
```
