```python
import numpy as np
import tensorflow as tf
from sklearn import discriminant_analysis
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

from sklearn.decomposition import PCA
```

```python
def plot_embedding(X, title=None):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    plt.figure()
    ax = plt.subplot(111)
    for i in range(X.shape[0]):
        plt.text(X[i, 0], X[i, 1], str(label_train[i]),
                 color=plt.cm.Set1(label_train[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    plt.xticks([]), plt.yticks([])
    if title is not None:
        plt.title(title)
```

```python
X_train = np.load('./data/train_doc.npy')
y_train = np.load('./data/train_label.npy')
label_train = np.nonzero(y_train)[1]
X_test = np.load('./data/test_doc.npy')
y_test = np.load('./data/test_label.npy')
label_test = np.nonzero(y_test)[1]

```

```python
lda = discriminant_analysis.LinearDiscriminantAnalysis(n_components=2)
X_train_lda = lda.fit_transform(X_train, label_train)
plt.figure(figsize=(10,10))
plt.scatter(X_train_lda[:,0], X_train_lda[:,1], s=4, c=label_train,  cmap='hsv')
```

```python
X_test_lda = lda.fit_transform(X_test, label_test)
plt.figure(figsize=(10,10))
plt.scatter(X_test_lda[:,0], X_test_lda[:,1], s=4, c=label_test,  cmap='hsv')
```
